```python
# Before running this code, make sure the Memo Directory is empty 
# and there exists directories Training and Test with the divided data

import os
from nltk.util import ngrams

n = 3

#Initialise System Paths
rootpath = "/mnt/E/Yandex/IITG/CS243 SPL/ADFA-LD/ADFA-LD"
processed_root = rootpath + "/Processed"
training_set_path = processed_root + "/Training"
test_set_path = processed_root + "/Test"
memo_path = processed_root + "/Memo"

#Training ngrams of various types
all_ngrams = {
    "Java_Meterpreter" : [],
    "Meterpreter" : [],
    "Web_Shell" : [],
    "Adduser" : [],
    "Hydra_FTP" : [],
    "Hydra_SSH" : [],
    "Normal" : [],
             }
freq_ngrams = {
    "Java_Meterpreter" : [],
    "Meterpreter" : [],
    "Web_Shell" : [],
    "Adduser" : [],
    "Hydra_FTP" : [],
    "Hydra_SSH" : [],
    "Normal" : [],
             }
top30_ngrams = []
training_types = ["Java_Meterpreter","Meterpreter","Web_Shell","Adduser","Hydra_FTP","Hydra_SSH","Normal"]
```

```python
def ngrams(inlist, n):
    
    out = []
    for i in range(len(inlist)-n+1):
        out.append(tuple(inlist[i:i+n]))
        
    return out
```

```python
def save_ngrams(content, dir_name):
    x = map(int, content.split()) #String to Integer list
    
    #Using Library
    #current_ngrams = ngrams(x, n, pad_left=False, pad_right=False, 
    #       left_pad_symbol=None, right_pad_symbol=None) #Find ngrams
    
    #Using Custom Code (Same asymptotic time complexity)
    current_ngrams = ngrams(x,n);
    
    for tt in training_types:
        if tt in dir_name:
            all_ngrams[tt].extend(current_ngrams) #Save to List
    
```

```python
def capture(arg, dir_name, files):
    '''
    capture: Function is invoked on each directory of the data set
    during directory walk
    
    arg - any specific arguments (same as that passed to os.path.walk)
    (in this implementation stores the name of the root directory on which walk was invoked)
    dire_name - name of the directory being traversed
    files - files inside the directory
    '''
    
    #Open File to which the files will be concatenated
    concat_file = open(memo_path+'/concat.txt','a')
    
    #if dir is not root
    if(arg!=dir_name):
        for f in files:
            #Go through each file and read it
            fullpath = dir_name + '/' + f
            data_file = open(fullpath,'r')
            content = data_file.read()
            data_file.close()
            
            #Save the ngrams
            save_ngrams(content,dir_name)
            
            #Write the content to concat.txt
            concat_file.write("\n# " + f + "\n" + content + "\n")
            
    concat_file.close()
    
    #print dir_name,files            
```

```python
arg = training_set_path
os.path.walk(training_set_path,capture,arg)
```

```python
from collections import Counter
# A Counter is a dict subclass for counting hashable objects. 
# It is an unordered collection where elements are stored as dictionary keys 
# and their counts are stored as dictionary values. 
# Counts are allowed to be any integer value including zero or negative counts.

for tt in training_types:
    f = open(memo_path + "/" + tt + ".txt",'w')
    curr_ngrams = sorted(all_ngrams[tt])
    
    freq = Counter(curr_ngrams)
    sorted_ngrams_by_freq = sorted(freq.items(),key=lambda x: x[1], reverse = True)
    
    f.write(''.join(str(e)+"\n" for e in sorted_ngrams_by_freq))
    f.close()
    freq_ngrams[tt].extend(sorted_ngrams_by_freq)
```

```python
import math

for tt in training_types:
    f = open(memo_path + "/" + tt + "_top30.txt",'w')
    curr_ngrams = freq_ngrams[tt]
    selected_indices = int(math.ceil(len(curr_ngrams)*0.3))
    top30 = curr_ngrams[:selected_indices]
    
    f.write(''.join(str(e)+"\n" for e in top30))
    f.close()
    top30_ngrams.extend(map((lambda x: x[0]),top30))
```

```python
unique_top30 = Counter(top30_ngrams).items()
unique_top30 = [x[0] for x in unique_top30]
sorted(unique_top30)
```

```python
def read_content(arg, dir_name, files):
    '''
    read_content: Function is invoked on each directory of the data set
    during directory walk
    
    arg - any specific arguments (same as he last arguement passed to os.path.walk)
    (in this implementation stores the path of file to write)
    dire_name - name of the directory being traversed
    files - files inside the directory
    '''
    caller_root = arg[0]
    writefile = arg[1]
    #if dir is not root
    if(caller_root!=dir_name):
        for f in files:
            #Go through each file and read it
            fullpath = dir_name + '/' + f
            data_file = open(fullpath,'r')
            content = data_file.read()
            data_file.close()
            
            #Save the ngrams
            save_feature_vector(content,dir_name,f,writefile)
    
    #print dir_name,files            
```
